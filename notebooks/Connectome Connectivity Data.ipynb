```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
def build_heatmap(title, csv_path):
    data = pd.read_csv(csv_path, header=None)
    data.index = ["schaefer.id" + str(i+1) for i in range(200)]
    data.columns = ["schaefer.id" + str(i+1) for i in range(200)]
    plt.figure(figsize=(20,15))
    g = sns.heatmap(data)
    g.set_yticklabels(g.get_yticklabels(), rotation=0)
    g.set_title(title)
    plt.tight_layout()
    plt.show()
```

```python
build_heatmap("Connectome", "Connectivity/Connectome_mean_rs3scaled_1065.csv")
```

```python
build_heatmap("Euclidean Distance", "Connectivity/Schaefer200_Euclidean_Distance_Matrix.csv")
```

```python
build_heatmap("Functional Connectivity", "Connectivity/N69_HC_functional_connectivity.csv")
```
